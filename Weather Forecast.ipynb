```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
le0= LabelEncoder()
le1 = LabelEncoder()
train_set['Precip Type']=le1.fit_transform(train_set['Precip Type'])
train_set['Summary']=le.fit_transform(train_set['Summary'])
train_set['Daily Summary']=le0.fit_transform(train_set['Daily Summary'])
train_set.head()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
train_set[train_set.columns[1:-1]] = scaler.fit_transform(train_set[train_set.columns[1:-1]])
train_set.head()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
test_set[test_set.columns[1:-1]] = scaler.fit_transform(test_set[test_set.columns[1:-1]])
test_set.head()
```

```python
date_column = test_set['Formatted Date']
```

```python
import numpy as np
from datetime import datetime

def discretize_date(current_date, t):
    current_date = current_date[:-10]
    cdate = datetime.strptime(current_date, '%Y-%m-%d %H:%M:%S')

    if t == 'hour_sin':
        return np.sin(2 * np.pi * cdate.hour/24.0)
    if t == 'hour_cos':
        return np.cos(2 * np.pi * cdate.hour/24.0)
    if t == 'day_sin':
        return np.sin(2 * np.pi * cdate.timetuple().tm_yday/365.0)
    if t == 'day_cos':
        return np.cos(2 * np.pi * cdate.timetuple().tm_yday/365.0)
```

```python
date_types = ['hour_sin', 'hour_cos', 'day_sin', 'day_cos']
for dt in date_types:
    train_set[dt] = train_set['Formatted Date'].apply(lambda x : discretize_date(x, dt))
train_set.drop(['Formatted Date'],axis=1,inplace=True)
```

```python
date_types = ['hour_sin', 'hour_cos', 'day_sin', 'day_cos']
for dt in date_types:
    test_set[dt] = test_set['Formatted Date'].apply(lambda x : discretize_date(x, dt))
test_set.drop(['Formatted Date'],axis=1,inplace=True)
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(max_depth=32,n_estimators=120,random_state=1)
model.fit(X,y)
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
```
